```python
df_fateh.category=df_fateh.category.str.lower()
df_fateh.category=[re.sub(r' +',r'_',c)+"---fateh" for c in df_fateh.category]

df_imtiaz.category=df_imtiaz.category.str.lower()
df_imtiaz.category=[re.sub(r' +',r'_',c)+"---imtiaz" for c in df_imtiaz.category]

df_jalalsons.category=df_jalalsons.category.str.lower()
df_jalalsons.category=[re.sub(r' +',r'_',c)+"---jalalsons" for c in df_jalalsons.category]

df_metro.category=df_metro.category.str.lower()
df_metro.category=[re.sub(r'-+',r'_',c)+"---metro" for c in df_metro.category]


df_fateh['common_category']=df_fateh['category'].map(category_mapping)
df_imtiaz['common_category']=df_imtiaz['category'].map(category_mapping)
df_jalalsons['common_category']=df_jalalsons['category'].map(category_mapping)
df_metro['common_category']=df_metro['category'].map(category_mapping)  
```
