```python
import pandas as pd
import re
```

```python
df_fateh=pd.read_csv("al-fateh-products-with-common-categories.csv")
df_imtiaz=pd.read_csv("Imtiaz-products-with-common-categories.csv")
df_jalalsons=pd.read_csv("jalal-sons-products-with-common-categories.csv")
df_metro=pd.read_csv("metro-products-with-common-categories.csv")
```

```python
common_categories_from_all_stores = set(df_fateh['common_category']).intersection(set(df_imtiaz['common_category'])).intersection(set(df_jalalsons['common_category'])).intersection(set(df_metro['common_category']))

all_common_categories = set(df_fateh['common_category']).union(set(df_imtiaz['common_category'])).union(set(df_jalalsons['common_category'])).union(set(df_metro['common_category']))

print("Number of common categories across all stores:")
print(len(common_categories_from_all_stores))
print("Number of all categories across the stores:")
print(len(all_common_categories))

print("Common categories across all stores:")
print(common_categories_from_all_stores)
print("Categories not present in all stores:")
print(all_common_categories.difference(common_categories_from_all_stores))
```

```python
common_fateh=df_fateh[df_fateh['common_category'].isin(common_categories_from_all_stores)]
a=common_fateh.groupby('common_category').agg(length=("name","count"))

b=df_fateh[df_fateh['common_category'].isin(all_common_categories.difference(common_categories_from_all_stores))].groupby('common_category').agg(length=("name","count"))

print("Categories present in all stores:")
print(a)
print("Categories not present in all stores:")
print(b)
```

```python
common_jalalsons=df_jalalsons[df_jalalsons['common_category'].isin(common_categories_from_all_stores)]
a=common_jalalsons.groupby('common_category').agg(length=("name","count"))

b=df_jalalsons[df_jalalsons['common_category'].isin(all_common_categories.difference(common_categories_from_all_stores))].groupby('common_category').agg(length=("name","count"))

print("Categories present in all stores:")
print(a)
print("Categories not present in all stores:")
print(b)
```

```python
common_metro=df_metro[df_metro['common_category'].isin(common_categories_from_all_stores)]
a=common_metro.groupby('common_category').agg(length=("name","count"))
b=df_metro[df_metro['common_category'].isin(all_common_categories.difference(common_categories_from_all_stores))].groupby('common_category').agg(length=("name","count"))

print("Categories present in all stores:")
print(a)
print("Categories not present in all stores:")
print(b)
```

```python
    
common_imtiaz=df_imtiaz[df_imtiaz['common_category'].isin(common_categories_from_all_stores)]
a=common_imtiaz.groupby('common_category').agg(length=("name","count"))

b=df_imtiaz[df_imtiaz['common_category'].isin(all_common_categories.difference(common_categories_from_all_stores))].groupby('common_category').agg(length=("name","count"))

print("Categories present in all stores:")
print(a)
print("Categories not present in all stores:")
print(b)

```

```python
df=pd.concat([common_fateh, common_jalalsons, common_metro, common_imtiaz],ignore_index=True)
df.to_csv("common-products-from-all-stores.csv",index=False)
```
